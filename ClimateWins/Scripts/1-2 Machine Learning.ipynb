## Checking for missing values

```python
# Use the isnull() method to check for missing values in each column, then sum the results
missing_values = df_climate.isnull().sum()
# Convert the missing_values Series to a DataFrame for better visualization
missing_values_df = pd.DataFrame(missing_values, columns=['Missing_Values_Count'])
# Display the DataFrame with the count of missing values for each column
print(missing_values_df)
```

## Getting list of column names

```python
# Use the columns attribute of the DataFrame to get the column names
columns = df_climate.columns.tolist()
```

```python
columns
```

```python
# Extract unique locations from the column names
locations = df_climate.columns.str.split('_').str[0].unique()

# Create a dictionary to store column names for each location
column_names_by_location = {}
for location in locations:
    columns_for_location = [col for col in df_climate.columns if col.startswith(location+'_')]
    column_names_by_location[location] = columns_for_location

# Display the dictionary of column names for each location
print(column_names_by_location)
```

```python
# Extract unique variable names from the column names so that variables can be grouped, such as all the temp_mean together or all of the humidity together
variables = set([col.split('_')[1] for col in df_climate.columns if '_' in col])

column_names_by_variable = {}
for variable in variables:
    columns_for_variable = [col for col in df_climate.columns if col.endswith('_'+variable)]
    column_names_by_variable[variable] = columns_for_variable

print(column_names_by_variable)
```

## Scaling the Data

```python
#create scaler object 
scaler = StandardScaler()
```

**Creating a scaled version with the column names from the original df_climate**

```python
#create a new df with the scaled data with the column names of the original df
df_ClimateScaled = pd.DataFrame(scaler.fit_transform(df_climate), columns=columns)
df_ClimateScaled.head()
```

**Creating a scaled version without the added column names from the original df_climate**

```python
#creating a scaled df, but without the column titles from the original 
df_scaled = pd.DataFrame(scaler.fit_transform(df_climate))
df_scaled.head()
```

```python
# Export dataframe as .pkl
df_ClimateScaled.to_pickle(os.path.join(path, 'DataSet_scaled.pkl'))
```

## Create subsets for variables for future analysis from scaled df

```python
# For df_ClimateScaled if column_names_by_variable is the dictionary of column names for each variable
# Create an empty dictionary to store subsets for each variable
subsets_by_variable = {}

# Loop through each variable and create subsets
for variable in column_names_by_variable:
    selected_columns = ['id', 'DATE', 'MONTH'] + column_names_by_variable[variable]
    subset_for_variable = df_ClimateScaled[selected_columns]
    subsets_by_variable[variable] = subset_for_variable

# Display the subsets for each variable
for variable, subset in subsets_by_variable.items():
    print(f"Subset for {variable}:")
    print(subset)
    print("\n")
```

## Create subsets for locations for future analysis from scaled df

```python
# column_names_by_location is the dictionary of column names for each location
# Create an empty dictionary to store subsets for each location
subsets_by_location = {}

# Loop through each location and create subsets
for location in column_names_by_location:
    selected_columns = ['id', 'DATE', 'MONTH'] + column_names_by_location[location]
    subset_for_location = df_ClimateScaled[selected_columns]
    subsets_by_location[location] = subset_for_location

# Display the subsets for each location
for location, subset in subsets_by_location.items():
    print(f"Subset for {location}:")
    print(subset)
    print("\n")
```

## Export subsets to csv files

```python
# subsets_by_location is the dictionary of subsets for each location
# Export each subset to a separate .csv file
for location, subset in subsets_by_location.items():
    file_name = f"{location}_subset.csv"
    file_path = os.path.join(path, 'Data', 'Prepared Data', file_name)
    subset.to_csv(file_path, index=False)
    print(f"Subset for {location} has been exported to {file_path}")
```

```python
# subsets_by_variable is the dictionary of subsets for each variable
# Export each variable subset to a separate .csv file in the specified directory
for variable, subset in subsets_by_variable.items():
    file_name = f"{variable}_subset.csv"
    file_path = os.path.join(path, 'Data', 'Prepared Data', file_name)
    subset.to_csv(file_path, index=False)
    print(f"Subset for {variable} has been exported to {file_path}")
```
